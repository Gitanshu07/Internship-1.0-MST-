```python
null_val = df.isnull().sum()/df.shape[0]*100
null_val
```

```python
missing_value_clm_gre_20 = null_val[null_val>20].keys()
missing_value_clm_gre_20
```

```python
df2_drop_clm = df.drop(columns=missing_value_clm_gre_20)
df2_drop_clm.shape
```

```python
df3_num = df2_drop_clm.select_dtypes(include=['int64', 'float64'])
df3_num.head()
```

```python
sns.heatmap(df3_num.isnull())
```

```python
df3_num[df3_num.isnull().any(axis=1)]
```

```python
df3_num.isnull().sum()
```

```python
missing_num_var = [var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
missing_num_var
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(missing_num_var):
    plt.subplot(2,2, i+1)
    sns.distplot(df3_num[var], bins = 20, kde_kws = {'linewidth': 5, 'color': '#DC143C'})
```

```python
# Normal distribution = mean,median,mode use hote hai 
# skewed distribution/asymetric - median use krte hai
```

```python
df4_num_mean = df3_num.fillna(df3_num.mean())
df4_num_mean.isnull().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(missing_num_var):
    plt.subplot(2,2, i+1)
    sns.distplot(df3_num[var], bins = 20, kde_kws = {'linewidth': 5, 'color': 'red'}, label='Origional')
    sns.distplot(df4_num_mean[var], bins = 20, kde_kws = {'linewidth': 5, 'color': 'green'}, label='Mean')
```

```python
df5_num_median = df3_num.fillna(df3_num.median())
df5_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(15,15))
sns.set()
for i,var in enumerate(missing_num_var):
    plt.subplot(2,2, i+1)
    sns.distplot(df3_num[var], bins = 20, hist=False, kde_kws = {'linewidth': 5, 'color': 'red'}, label='Original')
    sns.distplot(df4_num_mean[var], bins = 20, hist=False,kde_kws = {'linewidth': 3, 'color': 'green'}, label='Mean')
    sns.distplot(df5_num_median[var], bins = 20, hist=False, kde_kws = {'linewidth': 3, 'color': 'blue'}, label='Median')
plt.show()
```

```python

sns.set()
for i,var in enumerate(missing_num_var):
    plt.figure(figsize=(15,15))
    plt.subplot(3,1,1)
    sns.boxplot(df[var])
    plt.subplot(3,1,2)
    sns.boxplot(df4_num_mean[var])
    plt.subplot(3,1,3)
    sns.boxplot(df5_num_median[var])
plt.show()
```

```python
df_concat = pd.concat([df[missing_num_var], df4_num_mean[missing_num_var], df5_num_median[missing_num_var]], axis=1)
```
