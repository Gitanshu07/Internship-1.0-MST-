## Calculating Percentage of missing values

```python
null_var = data.isnull().sum()/data.shape[0]*100
null_var
```

```python
drop_columns = null_var[null_var>17].keys()
drop_columns
```

```python
data_drop_clm = data.drop(columns = drop_columns)
data_drop_clm.shape
```

```python
sns.heatmap(data_drop_clm.isnull())
```

## Selecting categorical values in this step

```python
cat_var = data_rows.select_dtypes(include = ['object']).columns
cat_var
```

```python
pd.concat([data['MSZoning'].value_counts()/data.shape[0]*100,
           data_rows['MSZoning'].value_counts()/data_rows.shape[0]*100],axis = 1,
         keys = ['MSZoning_org','MSZoning_clean'])
```

```python
def cat_var_dist(var):
    return pd.concat([data[var].value_counts()/data.shape[0]*100,
           data_rows[var].value_counts()/data_rows.shape[0]*100],axis = 1,
         keys = [var+' org',var+' clean'])
    
    
```

```python
cat_var_dist('LandSlope')
```

```python
plt.figure(figsize = (25,25))
for i,var in enumerate(cat_var):
    plt.subplot(19,2,i+1)
    sns.catplot(data[var],bins = 20)
    sns.catplot(data_rows[var],bins = 20)
```
