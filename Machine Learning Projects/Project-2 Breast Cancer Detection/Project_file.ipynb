```python
x = cancer_df.drop(['target'],axis = 1)
x.head()
```

```python
y = cancer_df['target']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
x_train.head()
```

```python
x_test.head()
```

## Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

## Model Training 

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

## Support Vector Classifier 

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='rbf')
svc.fit(x_train_sc,y_train)
```

```python
y_pred_1 = svc.predict(x_test_sc)
y_pred_1
```

```python
accuracy_score(y_test,y_pred_1)
```

## Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
lr_classifier = LogisticRegression(penalty='l2',C=1.0)
```

```python
lr_classifier.fit(x_train_sc,y_train)
```

```python
y_pred_2 = lr_classifier.predict(x_test_sc)
y_pred_2
```

```python
accuracy_score(y_test,y_pred_2)
```

## K - Nearest Neighbor Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
knn_classifier.fit(x_train_sc,y_train)
```

```python
y_pred_3 = knn_classifier.predict(x_test_sc)
y_pred_3
```

```python
accuracy_score(y_test,y_pred_3)
```

## Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb_classifier = GaussianNB()
nb_classifier.fit(x_train_sc,y_train)
```

```python
y_pred_4 = nb_classifier.predict(x_test_sc)
y_pred_4
```

```python
accuracy_score(y_test,y_pred_4)
```

## Decision Tree Classifier 

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dst_classifier = DecisionTreeClassifier(criterion='gini', random_state=51)
dst_classifier.fit(x_train,y_train)
```

```python
y_pred_5 = dst_classifier.predict(x_test)
y_pred_5
```

```python
accuracy_score(y_test,y_pred_5)
```
