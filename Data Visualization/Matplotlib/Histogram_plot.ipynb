```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
ml_students_age = np.random.randint(18,45,(100))
py_students_age = np.random.randint(18,40,(100))
print(ml_students_age)
print(py_students_age)
```

```python
plt.hist(ml_students_age)
plt.title("Machine learning Students")
plt.xlabel("Students age category")
plt.ylabel("No. of students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (16,9))
plt.hist([ml_students_age,py_students_age],bins,rwidth = 0.8,histtype='bar'
         ,orientation='vertical',color= ['r','g'],label = ["ML_Student","Py_Student"])
#plt.hist(py_students_age,bins,rwidth = 0.8,histtype='bar'
#         ,orientation='vertical',color= 'g',label = "Py_Student")
plt.title("Machine learning & Python Students")
plt.xlabel("Students age category")
plt.ylabel("No. of students age")
plt.legend()
plt.show()
```
