```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('googleplaystore.csv',nrows = 1000)
data.shape
data.head()
```

```python
x = data["Rating"]
y = data["Reviews"]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```
