```python
gw_df = gw_df.drop(columns = ["Country Code","Indicator Name","Indicator Code"],axis = 1).set_index("Country Name")
```

```python
plt.figure(figsize = (16,9))
sns.heatmap(gw_df)
plt.show()
```

```python
annot_arr = np.array([['a00','a01','a02'],['a10','a11','a12'],['a20','a21','a22'],['a30','a31','a32']])
annot_arr
```

```python
sns.heatmap(arr_2d ,annot=annot_arr ,fmt = "s")# S is used for sting type values
```

```python
plt.figure(figsize = (16,9))
annot_kws = {"fontsize":10,
            "fontstyle":"italic",
            'color':'k',
            'alpha':0.9,
            'verticalalignment':'center',
            "backgroundcolor":'w'}
sns.heatmap(gw_df, cmap = "coolwarm",annot = True,annot_kws=annot_kws)
```

```python
plt.figure(figsize = (16,9))
sns.heatmap(gw_df, cmap = "coolwarm",annot = True,linewidth=2,linecolor='k')
```

```python
plt.figure(figsize = (16,9))
sns.heatmap(gw_df, cmap = "coolwarm",annot = True,linewidth=2,linecolor='k',cbar=False)
```

```python
plt.figure(figsize = (16,9))
sns.heatmap(gw_df, cmap = "coolwarm",annot = True,linewidth=2,linecolor='k',cbar=False,xticklabels=False,yticklabels=False)
```

```python
plt.figure(figsize = (16,9))
cbar_kws = {"orientation":"horizontal", 
            "shrink":1,
            'extend':'min',
            'extendfrac':0.1,
            'ticks':np.arange(0,22),
            "drawedges":True }
sns.heatmap(gw_df, cmap = "coolwarm",annot = True,cbar_kws=cbar_kws)
```
