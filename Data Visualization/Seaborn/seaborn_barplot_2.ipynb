```python
tips_df = sns.load_dataset("tips")
```

```python
sns.barplot(x = tips_df.total_bill)
```

```python
sns.barplot(x = tips_df.day ,y= tips_df.total_bill)
```

```python
sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df)
```

```python
order = ['Sun','Thur','Fri','Sat']
sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,order = order)
```

```python
hue_order = ['Female','Male']
sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,hue_order = hue_order,estimator=np.mean)
```

```python
sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,hue_order = hue_order,estimator=np.max)
```

```python
sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,hue_order = hue_order,estimator=np.max,ci= 15)
```

```python
sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,hue_order = hue_order,estimator=np.max,
           n_boot = 3)
```

```python
sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,orient = 'v')
```

```python
sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,palette='magma')
```

```python
sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,palette='magma',saturation = 0.4)
```

```python
sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,palette='magma',errcolor='0.8')
```

```python
sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,palette='magma',errwidth=12)
```

```python
sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,palette='magma',dodge = False)
```

```python
kwargs = {'alpha':0.9,'linestyle':':','linewidth':5,'edgecolor':'k'}
sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,palette='magma',saturation = 0.4,
           **kwargs)
```

```python
sns.set()
kwargs = {'alpha':0.9,'linestyle':':','linewidth':5,'edgecolor':'k'}
sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,palette='magma',saturation = 0.4,
           **kwargs)
```

```python
sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,palette='magma',
            alpha=0.9,linestyle = '-.',edgecolor ='k')
```

```python
ax = sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,palette='magma',
            alpha=0.9,linestyle = '-.',edgecolor ='k')
ax.set(title = "Barplot of Tips DataFrame",
       xlabel = "Days",ylabel = "Total Bill")
```

```python
plt.figure(figsize = (16,9))
ax = sns.barplot(x = 'day' ,y= 'total_bill',hue = 'sex',
           data = tips_df,palette='magma',
            alpha=0.9,linestyle = '-.',edgecolor ='k')
ax.set(title = "Barplot of Tips DataFrame",
       xlabel = "Days",ylabel = "Total Bill")
plt.show()
```
