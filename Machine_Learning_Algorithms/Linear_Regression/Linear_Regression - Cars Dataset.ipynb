```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('cars_clus.csv', nrows=20)
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
drop = data[['manufact','model','type','partition','price','length','curb_wgt','fuel_cap','mpg','lnsales', 'resale', 'sales', 'horsepow', 'wheelbas', 'width']]
X = data.drop(drop,axis = 1)
X.shape
```

```python
Y = data['price']
Y.head()
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,Y)
```

```python
pred_y = lr.predict(X)
pred_y
```
