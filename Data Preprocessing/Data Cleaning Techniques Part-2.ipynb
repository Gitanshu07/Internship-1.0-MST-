```python
plt.figure(figsize = (25,20))
sns.heatmap(data.isnull())
```

```python
null_var = data.isnull().sum()/data.shape[0]*100
null_var
```

```python
drop_columns = null_var[null_var>20].keys()
drop_columns
```

```python
data_drop_clm = data.drop(columns = drop_columns)
data_drop_clm.shape
```

```python
data_row = data_drop_clm.select_dtypes(include = ['int64','float64'])
data_row.head()
```

```python
plt.figure(figsize = (25,20))
sns.heatmap(data_row.isnull())
```

```python
data_row[data_row.isnull().any(axis = 1)]
```

```python
data_row.isnull().sum()
```

```python
missing_var = [var for var in data_row.columns if data_row[var].isnull().sum()>0]
missing_var
```

```python
plt.figure(figsize = (25,25))
for i,var in enumerate(missing_var):
    plt.subplot(2,2,i+1)
    sns.distplot(data_row[var],bins = 20,kde_kws={'linewidth':5,'color':'#DC143C'})
    
```

```python
data_mean = data_row.fillna(data_row.mean())
data_mean.isnull().sum()
```

```python
plt.figure(figsize = (25,25))
sns.set()
for i,var in enumerate(missing_var):
    plt.subplot(2,2,i+1)
    sns.distplot(data_row[var],bins = 20,kde_kws={'linewidth':5,'color':'red'},label = 'Origional')
    sns.distplot(data_mean[var],bins = 20,kde_kws={'linewidth':5,'color':'green'},label= 'Mean value')
    plt.legend()
```

```python
data_median = data_row.fillna(data_row.median())
data_median.isnull().sum()
```

```python
plt.figure(figsize = (25,25))
sns.set()
for i,var in enumerate(missing_var):
    plt.subplot(2,2,i+1)
    sns.distplot(data_row[var],bins = 20,kde_kws={'linewidth':5,'color':'red'},label = 'Origional')
    sns.distplot(data_mean[var],bins = 20,kde_kws={'linewidth':5,'color':'green'},label= 'Mean value')
    sns.distplot(data_median[var],bins = 20,kde_kws={'linewidth':5,'color':'blue'},label= 'Median value')
    plt.legend()
```

```python
data_mode = data_row.fillna(data_row.mode())
data_mode.isnull().sum()
```

```python
plt.figure(figsize = (25,25))
sns.set()
for i,var in enumerate(missing_var):
    plt.subplot(2,2,i+1)
    sns.distplot(data_row[var],bins = 20,kde_kws={'linewidth':8,'color':'red'},label = 'Origional')
    sns.distplot(data_mean[var],bins = 20,kde_kws={'linewidth':5,'color':'green'},label= 'Mean value')
    sns.distplot(data_median[var],bins = 20,kde_kws={'linewidth':4,'color':'blue'},label= 'Median value')
    sns.distplot(data_mode[var],bins = 20,kde_kws={'linewidth':4,'color':'orange'},label= 'mode value')
    plt.legend()
```
